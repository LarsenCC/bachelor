# DATA PREPROCESSING


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
df = pd.read_excel("data/baza_HR_v5.xlsx")
df.head()
```

```python
df_clear = df.drop(['VO2_Mayers', 'W_step1_M_CF', 'W_step1_F_CF', 'W_step1_M_PW', 'W_step1_F_PW', 'W_step2_M', 'W_step2_F',
'W_step3_M_LT', 'W_step3_M_GT', 'W_step3_M_EQ', 'W_step3_F_LT', 'W_step3_F_GT', 'W_step3_F_EQ',  'W_step3_M',
'W_step3_F', 'W_step4_M', 'W_step4_F', 'Wasserman_vo2_l_kg_min', 'HRmax1', 'HRmax2'],axis=1)
df_clear.head()
```

```python
num_rows = len(df_clear.index)
num_rows
```

```python
print('Number of rows for max_hr:', len(df_clear['max_hr'].dropna()))
print('Number of rows for max_rer:', len(df_clear['max_rer'].dropna()))
print('Number of rows for vo2_l_min:', len(df_clear['vo2_l_min'].dropna()))
print('Number of rows for vo2_ml_kg_min:', len(df_clear['vo2_ml_kg_min'].dropna()))
```

```python
df_clear[df_clear['max_hr'].isna()]
```

```python
# Correlation plot ... just for overview
# plt.figure(figsize=(24,24))
# sns.heatmap(df_clear.corr(),cmap='coolwarm')
```

```python
# What to do with these values that have more than 95% of NaN values? DROP COLUMN?
# How do we fill in the rest? Median, Mean, 0?
data = []
rang = []
for el in df_clear.columns:
  num_of_elements = len(df_clear[el].dropna())
  percent = math.floor(num_of_elements/num_rows * 100)
  # print(el, "-->", percent)
  rang.append((el, percent))
  data.append(percent)
```

```python
rang.sort(key=lambda tup: tup[1])  # sorts in place
for i in range(0, len(rang)):
  el = rang[i][0]
  percent = rang[i][1]
  # print(i)
  print("{a:25s} --> {p:3d}".format(a = el, p = percent))
```

```python
plt.figure(figsize=(12,8))
sns.distplot(data, bins=100, kde=False)
```

GET DUMMY VARIABLES


```python
to_dummies = ['CR_Code', 'ethnic_code', 'Gender', 'PAStatus', 'Smoked ', 
              'educlevel', 'employment_status', 'marital_status', 
              'DiureticMed', 'PackDay', 'metabolic_system', 
              'protocol', 'Symptoms', 'test_termination_reason', 
              'TestIndication']

df_dummies = df_clear

for i in to_dummies:
    dummies = pd.get_dummies(df_clear[i], drop_first=True, dummy_na=True, prefix=i)
    start_with = str(i + '_')
    dummies.loc[df_dummies[i].isnull(), dummies.columns.str.startswith(start_with)] = np.nan
    dummies = dummies.drop(str(i + '_nan'), axis=1)
    df_dummies = df_dummies.drop(i, axis=1)
    df_dummies = pd.concat([df_dummies, dummies], axis=1)
```

```python
df_dummies.describe()
```

```python
df_dummies.head()
```

```python
data = []
rang = []
for el in df_dummies.columns:
    num_of_elements = len(df_dummies[el].dropna())
    percent = math.floor(num_of_elements/num_rows * 100)
    # print(el, "-->", percent)
    rang.append((el, percent))
    data.append(percent)

rang.sort(key=lambda tup: tup[1])  # sorts in place
for i in range(0, len(rang)):
    el = rang[i][0]
    percent = rang[i][1]
    # print(i)
    print("{a:30s} --> {p:3d}".format(a = el, p = percent))
```

```python
plt.figure(figsize=(12,8))
print('number of atributes:', len(data))
sns.distplot(data, bins=100, kde=False)
```

```python
testing = df_dummies[['@1rsSmoked', 'PackDay_1.0', 'PackDay_3.0', 'Smoked _1.0', 'Smoked _2.0']]
```

```python
# '@1rsSmoked', 'PackDay_1.0', 'PackDay_3.0' represent 2% or less ... so we will
# remove these attributes ...
# in addition, we will also concat 'Smoked _1.0', 'Smoked _2.0' into one
# variable, smoked or not ...
testing['Smoked'] = df_clear.loc[:,'Smoked ']
testing[testing['Smoked'] == 2] = 1
testing = testing.drop(['@1rsSmoked', 'PackDay_1.0', 'PackDay_3.0', 'Smoked _1.0', 'Smoked _2.0'], axis=1)
```

```python
testing['Smoked'].unique()
```

```python
testing = pd.concat([testing, df_dummies[['PhysInactive', 'PAStatus_1.0', 'PAStatus_2.0', 'PAStatus_3.0', 'PAStatus_4.0', 'PAStatus_5.0', 'PAStatus_6.0']]], axis=1)
testing.head()
```

```python
# it seems that PAStatus_1.0 represents an active person ...
# and PAStatus_5.0 an Inactive person, since they highly correlate with PhysInactive
# when PAStatus_1.0 == 1, PhysInactive == 0
# when PAStatus_5.0 == 1, PhysInactive == 1
# when PAStatus_6.0 == 1, PhysInactive == 1
# the more active you are the lower the PAStatus number ... 1 most active, 
# 6 the least active ...
testing[(testing['PhysInactive'] == 1) & (testing['PAStatus_6.0'] == 1)]
```

```python
testing[(testing['PhysInactive'] == 1) & (testing['PAStatus_5.0'] == 1)]
```

```python
testing[(testing['PhysInactive'] == 0) & (testing['PAStatus_1.0'] == 1)]
```

```python
testing[(testing['PhysInactive'] == 0) & (testing['PAStatus_4.0'] == 1)]
```

```python
testing[(testing['PhysInactive'] == 0) & (testing['PAStatus_3.0'] == 1)]
```

```python
testing[(testing['PhysInactive'] == 0) & (testing['PAStatus_2.0'] == 1)]
```

```python
# it turns out, the best way to merge these attributes is by saying
# that PAStatus 5 and 6 are Physically Inactive people --> PhysInactive = 1
# PAStatus 1,2,3,4 are Physically Active people --> PhysInactive = 0
testing[testing['PAStatus_6.0'] == 1]['PhysInactive'] = 1
testing[testing['PAStatus_5.0'] == 1]['PhysInactive'] = 1
testing[testing['PAStatus_1.0'] == 1]['PhysInactive'] = 0
testing[testing['PAStatus_2.0'] == 1]['PhysInactive'] = 0
testing[testing['PAStatus_3.0'] == 1]['PhysInactive'] = 0
testing[testing['PAStatus_4.0'] == 1]['PhysInactive'] = 0
testing = testing.drop(['PAStatus_1.0', 'PAStatus_2.0', 'PAStatus_3.0', 'PAStatus_4.0', 'PAStatus_5.0', 'PAStatus_6.0'], axis=1)
```

```python
testing = pd.concat([testing, df_dummies[['resting_ecg', 'gxt_arr']]], axis=1)
testing.head()
```

IMPUTACIJA



```python
# Basic ... mean and median ...
testing.describe()
```

```python
for col in testing:
    if col == 'Numbers':
        continue
    d = testing[col].describe()
    replace = 0
  # print(d.loc['max'])
    if (d.loc['max'] == 1):
    # we have a binary attribute, so take median
        replace = d.loc['50%']
    else:
    # else we have continuous variable, so we take mean
        replace = round(d.loc['mean'], 2)
    print(replace)
    testing[col] = testing[col].fillna(replace)
```

```python
df_final = testing_scaled.merge(df_dummies[['max_hr', 'vo2_l_min', 'vo2_ml_kg_min', 'Numbers']], on='Numbers').drop('Numbers', axis=1).dropna()
# df_final
```

```python
# EXPORT
df_final.to_csv('basic2.csv')
```

```python
# Load some data and put it in a numpy.array matrix
my_input_matrix = np.array(df_final.drop(['max_hr', 'vo2_l_min', 'vo2_ml_kg_min'], axis=1).values)
```

```python
# max_hr
# Load the label vector
my_label_vector = np.array(df_final['max_hr'].values)

r_max_hr = relief.Relief(
    n_features=5 # Choose the best 5 features
) # Will run by default on all processors concurrently

my_transformed_matrix = r_max_hr.fit_transform(
    my_input_matrix,
    my_label_vector
)

print(r_max_hr.w_)
```

```python
# vo2_l_min
# Load the label vector
my_label_vector = np.array(df_final['vo2_l_min'].values)

r_vo2_l_min = relief.Relief(
    n_features=5 # Choose the best 5 features
) # Will run by default on all processors concurrently

my_transformed_matrix = r_vo2_l_min.fit_transform(
    my_input_matrix,
    my_label_vector
)

print(r_vo2_l_min.w_)
```

```python
# vo2_ml_kg_min
# Load the label vector
my_label_vector = np.array(df_final['vo2_ml_kg_min'].values)

r_vo2_ml_kg_min = relief.Relief(
    n_features=5 # Choose the best 5 features
) # Will run by default on all processors concurrently

my_transformed_matrix = r_vo2_ml_kg_min.fit_transform(
    my_input_matrix,
    my_label_vector
)

print(r_vo2_ml_kg_min.w_)
```

```python
columns = df_final.drop(['max_hr', 'vo2_l_min', 'vo2_ml_kg_min'], axis=1).columns
```

```python
# max_hr
attr = columns[r_max_hr.w_ > 0]
w = r_max_hr.w_[r_max_hr.w_ > 0]
attr = [x for _, x in sorted(zip(w,attr))]
w = np.sort(w)
for i in range(0, len(attr)):
  print("{a:25s} --> {p:3f}".format(a = attr[i], p = w[i]))
```

```python
# vo2_l_min
attr = columns[r_vo2_l_min.w_ > 0]
w = r_vo2_l_min.w_[r_vo2_l_min.w_ > 0]
attr = [x for _, x in sorted(zip(w,attr))]
w = np.sort(w)
for i in range(0, len(attr)):
  print("{a:25s} --> {p:3f}".format(a = attr[i], p = w[i]))
```

```python
# vo2_ml_kg_min
attr = columns[r_vo2_ml_kg_min.w_ > 0]
w = r_vo2_ml_kg_min.w_[r_vo2_ml_kg_min.w_ > 0]
attr = [x for _, x in sorted(zip(w,attr))]
w = np.sort(w)
for i in range(0, len(attr)):
  print("{a:25s} --> {p:3f}".format(a = attr[i], p = w[i]))
```
